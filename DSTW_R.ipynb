```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Load the Iris dataset
data(iris)
```

```r
# View the first few rows of the dataset
head(iris)

# Get the structure of the dataset
str(iris)

# Check for missing values
sum(is.na(iris))
```

```r
# Summary of the dataset
summary(iris)

# Check the unique values of the 'variety' column (species)
unique(iris$Species)
```

```r
# Pair plot to visualize relationships between numeric features
pairs(iris[, 1:4], col=iris$Species, pch=19, main="Pairwise Scatter Plots of Iris Dataset")
```

```r
# Boxplot of Sepal Length across different Species
ggplot(iris, aes(x=Species, y=Sepal.Length, fill=Species)) +
  geom_boxplot() +
  theme_minimal() +
  labs(title="Sepal Length Distribution by Species", x="Species", y="Sepal Length")
```

```r
# Histogram of Sepal Length
ggplot(iris, aes(x=Sepal.Length, fill=Species)) +
  geom_histogram(binwidth=0.2, alpha=0.7, position="identity") +
  theme_minimal() +
  labs(title="Histogram of Sepal Length", x="Sepal Length", y="Frequency")
```

```r
# Density plot for Petal Length
ggplot(iris, aes(x=Petal.Length, fill=Species)) +
  geom_density(alpha=0.7) +
  theme_minimal() +
  labs(title="Density Plot of Petal Length by Species", x="Petal Length", y="Density")
```

```r
# Bar plot of the count of each species
ggplot(iris, aes(x=Species, fill=Species)) +
  geom_bar() +
  theme_minimal() +
  labs(title="Count of Each Iris Species", x="Species", y="Count")
```

```r
install.packages("viridis")
```

```r
# Compute the correlation matrix of numeric features
cor_matrix <- cor(iris[, 1:4])

# Print the correlation matrix
print(cor_matrix)

# Heatmap of the correlation matrix
library(reshape2)
library(viridis)

cor_matrix_melted <- melt(cor_matrix)
ggplot(cor_matrix_melted, aes(Var1, Var2, fill=value)) +
  geom_tile() +
  scale_fill_viridis() +
  theme_minimal() +
  labs(title="Correlation Matrix Heatmap")
```

```r
# Scatter plot with a linear regression line between Sepal Length and Petal Length
ggplot(iris, aes(x=Sepal.Length, y=Petal.Length, color=Species)) +
  geom_point() +
  geom_smooth(method="lm", se=FALSE) +
  theme_minimal() +
  labs(title="Linear Relationship between Sepal Length and Petal Length", x="Sepal Length", y="Petal Length")
```

```r
# Perform PCA on the numeric features
pca <- prcomp(iris[, 1:4], center = TRUE, scale. = TRUE)

# Summary of PCA
summary(pca)

# Plot the first two principal components
pca_data <- data.frame(pca$x)
pca_data$Species <- iris$Species

ggplot(pca_data, aes(x=PC1, y=PC2, color=Species)) +
  geom_point(size=3) +
  theme_minimal() +
  labs(title="PCA Plot of Iris Dataset", x="PC1", y="PC2")
```
